```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
# Distribution
sns.distplot(data['Internet_users'])
```

```python
# Boxplot
ax = sns.boxplot(data = data, x='Income_Group', y='Birth_rate')
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
```

```python
# scatter plot with regression line
sns.lmplot(data = data, x = 'Internet_users', y = 'Birth_rate')
```

```python
# scatter plot with different groups
sns.lmplot(data = data, x = 'Internet_users', y = 'Birth_rate', fit_reg = False,hue = 'Income_Group')
```
